```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ðŸ“Œ Load the dataset
df = pd.read_csv("../data/heart.csv")  # Ensure the correct file path
```

```python
# ðŸ“Œ Display the first 5 rows
df.head()
```

```python
# ðŸ“Œ Get dataset overview
df.info()

# ðŸ“Œ Summary statistics
df.describe()

# ðŸ“Œ Check for missing values
df.isnull().sum()
```

```python
# ðŸ“Œ Drop missing values if any
df.dropna(inplace=True)

# ðŸ“Œ Remove duplicate records if any
df.drop_duplicates(inplace=True)

# ðŸ“Œ Verify data after cleaning
df.info()
```

```python
# Set style
sns.set_style("whitegrid")

# ðŸ“Œ Plot the age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20, kde=True, color="blue")
plt.title("Distribution of Patients' Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='target', y='age', data=df, hue='target', palette="coolwarm")
plt.title("Age Distribution by Heart Disease Presence")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.legend([],[], frameon=False)  # Hide duplicate legend
plt.show()

```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)
print("âœ… Cleaned dataset has been saved successfully!")
```
